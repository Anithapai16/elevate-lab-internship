so here we can see that there are outliers in the age column so we use median to fill the missing value

```python
df['Age']=df['Age'].fillna(df['Age'].median())
```

```python
df.drop('Cabin',axis=1,inplace=True)
#since more than 80% missing values we drop the column
```

```python
df['Embarked']=df['Embarked'].fillna(df['Embarked'].mode()[0])
```
